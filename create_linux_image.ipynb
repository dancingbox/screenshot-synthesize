```python
# Adds text to a desktop shortcut image (below the icon)
def addShortcutText(image, text, font, icon_within_border, icon_text_max_height, icon_image_max_height):
    # Convert the image to RGB (OpenCV uses BGR)  
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA)  

    # Pass the image to PIL  
    image_pil = Image.fromarray(image_rgb)  
    draw = ImageDraw.Draw(image_pil)  

    # use a truetype font  
    w, _ = draw.textsize(text, font=font)
    _, h = draw.textsize("h", font=font)

    # Get text coordinates
    textX = int((image.shape[1] - w) / 2)
    textY = int((icon_text_max_height - h) / 2) + icon_within_border
    
    # Draw the text
    draw.text((textX, textY + icon_image_max_height), text, font=font, fill=(255, 255, 255, 255))
    
    # Get back the image to OpenCV  
    image = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGBA2BGRA)
    
    return image
```

```python
# Adds text to a given background image
def createTextField(image, text, font, color, align_bottom=False):
    # Additional space to left and right of text
    left_border = 2
    right_border = 5
    
    # Make sure background image is large enough in terms of width
    while image.shape[1] < 500:
        image = np.hstack((image, image))
    
    # Convert the image to RGB (OpenCV uses BGR)  
    textImage_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

    # Pass the image to PIL  
    textImage_pil = Image.fromarray(textImage_rgb)
    draw = ImageDraw.Draw(textImage_pil)

    # Use a truetype font  
    w, _ = draw.textsize(text, font=font)
    _, h = draw.textsize("h", font=font)

    # Get text x-coordinate
    textX = left_border
    
    # get text y-coordinate
    if not align_bottom:
        textY = int((image.shape[0] - h) / 2) - 1
    else:
        textY = image.shape[0] - h - 10
    
    # Draw the text
    draw.text((textX, textY), text, font=font, fill=color)

    # Get back the image to OpenCV  
    textImage = cv2.cvtColor(np.array(textImage_pil), cv2.COLOR_RGB2BGR)
    
    # Fit image to text size (remove remaining background)
    textImage = textImage[0:textImage.shape[1], 0:w+left_border+right_border]
    
    return textImage
```

```python
# Overlays a transparent image onto another transparent image
def insertTransparentImage(image, overlay, x1, x2, y1, y2):
    img = image.copy()
    
    alpha_overlay = overlay[:, :, 3] / 255.0
    alpha_image = 1.0 - alpha_overlay

    # Intersect by alpha-value
    for c in range(0, 3):
        img[y1:y2, x1:x2, c] = (alpha_overlay * overlay[:, :, c] + alpha_image * img[y1:y2, x1:x2, c])
        
    return img
```

```python
# Overlays a transparent image onto another transparent without intersecting alpha values (alpha > 120 is used as mask)
def overlayTransparentImage(image, overlay, x1, x2, y1, y2):
    img = image.copy()
    
    for i in range(y1, y2):
        for j in range(x1, x2):
            # Check wheter we should take pixel from image or from overlay image
            if overlay[i-y1, j-x1, 3] > 120:
                img[i, j] = overlay[i-y1, j-x1]
        
    return img
```

```python
# Adds a border effect to a given image
def addWindowBorderEffect(window):
    # Get border size
    size = border_effect_size
    
    # Make image larger such that border fits
    image = np.zeros((window.shape[0]+size*2, window.shape[1]+size*2, 4))
    
    # Draw border line-by-line
    for i in range(0, size):
        # Get transparency of current line
        transparency = 255 / (2 ** (size-i))
        
        # Get border color
        color = (120, 120, 120, transparency)
        
        # Draw border line
        image[i:i+1, i:image.shape[1]-i] = color
        image[i:image.shape[0]-i, image.shape[1]-i-1:image.shape[1]-i] = color
        image[image.shape[0]-i-1:image.shape[0]-i, i:image.shape[1]-i] = color
        image[i:image.shape[0]-i, i:i+1] = color
    
    # Add window to the drawn border
    image[size:size+window.shape[0], size:size+window.shape[1]] = window
    
    return image
```

```python
# Create an empty desktop image
def createDesktop():
    # Copy left and right topbar ends
    topbar_start = topbar_start_original.copy()
    topbar_end = topbar_end_original.copy()
    
    # Select a random wallpaper
    if len(os.listdir(path_to_wallpapers)) == 0:
        print('Did not find any wallpapers in ' + path_to_wallpapers)
    wallpaper_file = random.choice(os.listdir(path_to_wallpapers))
    wallpaper = cv2.imread(os.path.join(path_to_wallpapers, wallpaper_file))
    
    # Only scales to width!
    # wallpaper = imutils.resize(wallpaper, width=wallpaper_width)
    wallpaper = cv2.resize(wallpaper, (wallpaper_width, wallpaper_height))

    # ==> Taskbar top (left):

    # Show application title in topbar in case there is at least one application open (which we assume)
    number_of_applications = 1
    if number_of_applications > 0:
        # Pick a random application icon
        icon_name = random.choice(os.listdir(path_to_left_topbar_elements))

        # Load and resize icon
        icon = cv2.imread(os.path.join(path_to_left_topbar_elements, icon_name), cv2.IMREAD_UNCHANGED)
        icon = imutils.resize(icon, height=topbar_start.shape[0])

        # Resize icon such that it doesn't fill whole topbar
        icon = imutils.resize(icon, height=int(topbar_start.shape[0] * 0.75))
        
        # Get background for icon image
        image = topbar_filler.copy()
        image = image[:, 0:icon.shape[1]]

        # Center icon
        x = int((image.shape[1] - icon.shape[1]) / 2)
        y = int((image.shape[0] - icon.shape[0]) / 2)

        # Draw icon on background
        insertTransparentImage(image, icon, x, x+icon.shape[1], y, y+icon.shape[0])
        
        # Append icon to left topbar element
        topbar_start = np.hstack((topbar_start, image))

        # Our title text vocabulary
        words = ["Lorem", "Ipsum", "Dolor", "Consetetur", "Sadipscing", "Elitr", "Nonumy", "Eirmod", "Tempor", "Invidunt",
                 "Labore", "Dolore", "Magna", "Aliquyam", "Erat", "Voluptua", "Vero", "Accusam", "Dolores", "Rebum"]
        
        # Create random application name
        text = " ".join(random.sample(words, k=random.randint(1,3)))

        # Get background for title text
        textImage = topbar_filler.copy()

        # Get font
        fontsize = int(textImage.shape[0] / 1.8)
        font = ImageFont.truetype(path_to_font, fontsize)  

        # Draw text on background
        textImage = createTextField(textImage, text, font, (255, 255, 255, 255))
        
        # Append application title text to topbar left
        topbar_start = np.hstack((topbar_start, textImage))
        
        # Append open application symbol to topbar left
        topbar_start = np.hstack((topbar_start, topbar_symbol))

    # Draw the left part of the topbar on the wallpaper
    x1 = 0
    x2 = x1 + topbar_start.shape[1]
    
    y1 = 0
    y2 = y1 + topbar_start.shape[0]

    wallpaper[y1:y2, x1:x2] = topbar_start

    # ==> Taskbar top (right):

    # Get random icons for the right part of the topbar
    number_of_elements = random.randint(2, len(os.listdir(path_to_right_topbar_elements)))
    icons = random.sample(os.listdir(path_to_right_topbar_elements), k=number_of_elements)

    # Append the random icons to the right part of topbar
    for i in range(0, len(icons)):
        icon = cv2.imread(os.path.join(path_to_right_topbar_elements, icons[i]))
        icon = imutils.resize(icon, height=topbar_end.shape[0])

        # Append icon
        topbar_end = np.hstack((icon, topbar_end))

    # Draw right part of topbar on the wallpaper
    x1 = wallpaper.shape[1] - topbar_end.shape[1]
    x2 = x1 + topbar_end.shape[1]
    
    y1 = 0
    y2 = y1 + topbar_end.shape[0]

    wallpaper[y1:y2, x1:x2] = topbar_end

    
    # ==> Taskbar top (middle):

    # Draw the clock in the middle of the topbar on the wallpaper
    x_offset_middle = int((wallpaper.shape[1] - topbar_middle.shape[1]) / 2)
    y_offset = 0

    wallpaper[y_offset:y_offset+topbar_middle.shape[0], x_offset_middle:x_offset_middle+topbar_middle.shape[1]] = topbar_middle
    
    # Fill up middle of topbar (left), since in the middle there is the clock
    fillwidth = x_offset_middle - topbar_start.shape[1]
    y_offset = 0

    # If there is still space we need to fill, then... (filler image is not infinite, so we may need to fill up multiple times)
    if fillwidth > 0:
        i = 0
        while i < fillwidth:
            # Get x-coordinate
            x_offset = topbar_start.shape[1] + i
            
            # Cut image if whole image would be larger than fillwidth
            width = min(topbar_filler.shape[1], fillwidth - i)

            # Add filler image
            wallpaper[y_offset:y_offset+topbar_filler.shape[0], x_offset:x_offset+width] = topbar_filler[:, 0:width]

            # Update fillwidth
            i += topbar_filler.shape[1]

    # Fill up middle of topbar (right), since in the middle there is the clock
    fillwidth = wallpaper.shape[1] - topbar_end.shape[1] - (x_offset_middle + topbar_middle.shape[1])
    y_offset = 0

    # If there is still space we need to fill, then... (filler image is not infinite, so we may need to fill up multiple times)
    if fillwidth > 0:
        i = 0
        while i < fillwidth:
            # Get x-coordinate
            x_offset = x_offset_middle + topbar_middle.shape[1] + i
            
            # Cut image if whole image would be larger than fillwidth
            width = min(topbar_filler.shape[1], fillwidth - i)

            # Add filler image
            wallpaper[y_offset:y_offset+topbar_filler.shape[0], x_offset:x_offset+width] = topbar_filler[:, 0:width]

            # Update fillwidth
            i += topbar_filler.shape[1]
            
            
    # ==> Taskbar left:
    
    # Pick random number of tasks for the taskbar
    number_of_elements = random.randint(3, len(os.listdir(path_to_taskbar_elements)))
    tasks = random.sample(os.listdir(path_to_icons), k=number_of_elements)

    # Start with empty top element
    taskbar_start = None

    # Append task icons from top to bottom
    for i in range(0, len(tasks)):
        # Load dot overlay effect
        dot_overlay = imutils.resize(taskbar_dot, height=taskbar_end.shape[0])
        
        # Create transparent task background (which has correct size)
        task_background = np.ones((taskbar_end.shape[0], taskbar_end.shape[1], 4))
        
        # Load and resize icon
        task = cv2.imread(os.path.join(path_to_icons, tasks[i]), cv2.IMREAD_UNCHANGED)
        task = imutils.resize(task, width=int(taskbar_end.shape[1] * 0.75))
        
        # Compute center to place the icon
        x1 = int(dot_overlay.shape[1]/2) + int((task_background.shape[1] - task.shape[1]) / 2)
        x2 = x1 + task.shape[1]
        
        y1 = int((task_background.shape[0] - task.shape[0]) / 2)
        y2 = y1 + task.shape[0]
        
        # Add task icon to transparent background with correct size
        task_background[y1:y2, x1:x2] = task
        task = task_background
            
        # Randomly decide if task has "task open" effect
        coinflip = random.randint(0, 4)
        if coinflip == 0:
            # Draw "task open" dot effect
            task = overlayTransparentImage(task_background, dot_overlay, 0, dot_overlay.shape[1], 0, dot_overlay.shape[0])
        
            # Randomly decide if we also add "Task selected" effect
            coinflip = random.randint(0, 3)
            
            if coinflip == 0:
                # Draw "Task selected" effect
                task = insertTransparentImage(taskbar_selected, task, 0, task.shape[1], 0, task.shape[0])

        # Append to taskbar
        if taskbar_start is None:
            taskbar_start = task
        else:
            taskbar_start = np.vstack((taskbar_start, task))
            
        # If there is not enough space to add another task to taskbar, abort
        if taskbar_start.shape[0] + task.shape[0] + taskbar_end.shape[0] + topbar_start.shape[0] > wallpaper.shape[0]:
            break
    
    # Get the transparent black background for the taskbar
    taskbar_background_normal = np.ones((wallpaper.shape[0] - topbar_start.shape[0], taskbar_end.shape[1], 4))
    taskbar_background_normal[:, :] = (70, 70, 70, 200)
    
    # Get the darker transparent black background for the taskbar in case something is in fullscreen
    taskbar_background_fullscreen = np.ones((wallpaper.shape[0] - topbar_start.shape[0], taskbar_end.shape[1], 4))
    taskbar_background_fullscreen[:, :] = (70, 70, 70, 240)

    # Draw taskbar icons on the two different backgrounds for the taskbar
    x1 = 0
    x2 = x1 + taskbar_start.shape[1]

    y1 = 0
    y2 = y1 + taskbar_start.shape[0]

    taskbar_background_normal = overlayTransparentImage(taskbar_background_normal, taskbar_start, x1, x2, y1, y2)
    taskbar_background_fullscreen = overlayTransparentImage(taskbar_background_fullscreen, taskbar_start, x1, x2, y1, y2)

    # Draw the last special taskbar element on the two different taskbar backgrounds
    x1 = 0
    x2 = x1 + taskbar_end.shape[1]

    y1 = taskbar_background_normal.shape[0] - taskbar_end.shape[0]
    y2 = y1 + taskbar_end.shape[0]

    taskbar_background_normal = overlayTransparentImage(taskbar_background_normal, taskbar_end, x1, x2, y1, y2)
    taskbar_background_fullscreen = overlayTransparentImage(taskbar_background_fullscreen, taskbar_end, x1, x2, y1, y2)

    # Return synthesized screenshot plus the two different possible taskbars
    return wallpaper, taskbar_background_normal, taskbar_background_fullscreen, wallpaper_file
```

```python
# Create a random window head for a given width
def generateWindowHead(final_width):
    # Copy the start window head element
    start = window_start.copy()

    # Our title vocabulary
    words = ["Lorem", "Ipsum", "Dolor", "Consetetur", "Sadipscing", "Elitr", "Nonumy", "Eirmod", "Tempor", "Invidunt",
             "Labore", "Dolore", "Magna", "Aliquyam", "Erat", "Voluptua", "Vero", "Accusam", "Dolores", "Rebum"]

    # Create a random title
    text = " ".join(random.sample(words, k=random.randint(1,3)))

    # Background image for title text
    textImage = window_middle.copy()

    #Get font
    fontsize = int(textImage.shape[0] / 1.8)
    font = ImageFont.truetype(path_to_font, fontsize)  

    # Draw title text on background image
    textImage = createTextField(textImage, text, font, (255, 255, 255, 255))

    # Fill up left part of title with filler image
    fillwidth = int((final_width - textImage.shape[1]) / 2) - start.shape[1]
    y_offset = 0

    if fillwidth > 0:
        # There is still space to fill, so fill
        i = 0
        while i < fillwidth:
            # Cut image if whole image would be larger than fillwidth
            width = min(window_middle.shape[1], fillwidth - i)
            
            # Append filler image
            start = np.hstack((start, window_middle[:, 0:width]))
            
            # Update remaining fillwidth
            i += window_middle.shape[1]
    else:
        # Application head is already too large. Cut such that at least left end still fits to it
        textImage = textImage[:, 0:(final_width-start.shape[1]-window_end.shape[1])]
    
    #Append the title, since now left part has been filled up
    textImage = cv2.cvtColor(textImage, cv2.COLOR_RGB2RGBA)
    start = np.hstack((start, textImage))
    
     # Fill up right part of title with filler image
    fillwidth = final_width - start.shape[1] - window_end.shape[1]
    y_offset = 0
    
    if fillwidth > 0:
        # There is still space to fill, so fill
        i = 0
        while i < fillwidth:
            # Cut image if whole image would be larger than fillwidth
            width = min(window_middle.shape[1], fillwidth - i)
            
            # Append filler image
            start = np.hstack((start, window_middle[:, 0:width]))
            
            # Update remaining fillwidth
            i += window_middle.shape[1]
    
    # Append end of application head
    start = np.hstack((start, window_end))
    
    return start

def generateBrowserHead(final_width):
    start = firefox_start.copy()
    
    # Get a random URL
    words = ["lorem", "ipsum", "dolor", "sit", "amet", "consetetur", "sadipscing", "elitr", "sed", 
             "diam", "nonumy", "eirmod", "tempor", "invidunt", "ut", "labore", "et", "dolore", "magna", 
             "aliquyam", "erat", "sed", "diam", "voluptua", "at", "vero", "eos", "et", "accusam", "et", 
             "justo", "duo", "dolores", "rebum"]

    # Seperator characters which connect the random URL words
    separators = ["/", ".", "-", "_", "=", "+", ""]
    
    # Create random URL
    samples = random.sample(words, k=random.randint(3,15))
    text = ""
    
    # Add random seperators between words
    for s in samples:
        sep = random.choice(separators)
        text += s
        text += sep
    
    # Get background for URL
    textImage = firefox_middle.copy()
    
    # Get URL font
    fontsize = int(textImage.shape[0] / 4)
    font = ImageFont.truetype(path_to_font, fontsize) 

    # Add URL text to background
    textImage = createTextField(textImage, text, font, (0, 0, 0, 255), align_bottom=True)
    
    # Append URL text to start image of browser head
    start = np.hstack((start, textImage))
    
    # Fill remaining width with filler image
    fillwidth = final_width - start.shape[1] - firefox_end.shape[1]
    y_offset = 0

    if fillwidth > 0:
        # There is still space to fill, so fill
        i = 0
        while i < fillwidth:
            # Cut image if whole image would be larger than fillwidth
            width = min(firefox_middle.shape[1], fillwidth - i)
            
            # Append filler image
            start = np.hstack((start, firefox_middle[:, 0:width]))
            
            # Update remaining fillwidth
            i += firefox_middle.shape[1]
    else:
        # Browser head is already too large. Cut such that at least left end still fits to it
        start = start[:, 0:(final_width-firefox_end.shape[1])]
    
    # Add right end of browser head
    start = np.hstack((start, firefox_end))
    
    return start
```

```python
# Add a random browser window to the desktop screenshot
def addBrowerWindow(screen, website_name, fullscreen):
    # Captures whether we take the large or small version of a website
    version = "large"
    
    # If application is not shown in fullscreen, randomly decide if we go for small or large version
    if not fullscreen:
        coinflip = random.randint(0, 1)
        
        if coinflip == 0:
            version = "small"
    
    # Load website screenshot
    website_path = os.path.join(path_to_websites, version, website_name)
    website = cv2.imread(website_path)
    
    if website is None:
        print(website_path)
    
    # Resize website such that it is not lager than the screenshot
    website = imutils.resize(website, width=min(screen.shape[1] - taskbar_end.shape[1] - border_effect_size * 2, website.shape[1]))
    website = imutils.resize(website, height=min(screen.shape[0] - topbar_start_original.shape[0] - window_end.shape[0] - firefox_end.shape[0] - border_effect_size * 2, website.shape[0]))

    # Disable fullscreen mode for webpages which would be rescaled too large
    if website.shape[0] < screen.shape[0] / 2.5 or website.shape[1] < screen.shape[1] / 2.5:
        #print(website_path, " is too small for fullscreen mode!")
        fullscreen = False
        
    if fullscreen:
        # Resize webpage to fullscreen
        website = cv2.resize(website, (screen.shape[1] - taskbar_end.shape[1], screen.shape[0] - topbar_start_original.shape[0] - window_start.shape[0] - firefox_start.shape[0]))
    else:
        # Randomly scale the webpage, since it's not in fullscreen
        scale = random.uniform(0.7, 1.0)
        website = imutils.resize(website, width=int(website.shape[1] * scale))
    
    # Generate a random application and browser head for webpage
    head = generateWindowHead(website.shape[1])
    browserHead = generateBrowserHead(website.shape[1])
    
    # Add alpha channels
    website = cv2.cvtColor(website, cv2.COLOR_RGB2RGBA)
    browserHead = cv2.cvtColor(browserHead, cv2.COLOR_RGB2RGBA)
    
    website = np.vstack((head, browserHead, website))
    
    # Add window effect to webpage (if not in fullscreen)
    if not fullscreen:
        website = addWindowBorderEffect(website)
    
    # Choose random position for webpage
    x1 = random.randint(taskbar_end.shape[1], screen.shape[1] - website.shape[1])
    x2 = x1 + website.shape[1]
    
    y1 = random.randint(topbar_start_original.shape[0], screen.shape[0] - website.shape[0])
    y2 = y1 + website.shape[0]
    
    # Add webpage to screenshot
    screen = insertTransparentImage(screen, website, x1, x2, y1, y2)
    
    data = {"name": website_path, "fullscreen": fullscreen, "x1": x1, "x2": x2, "y1": y1, "y2": y2}
    
    # return screenshot with added webpage and whether its in fullscreen mode or not
    return screen, fullscreen, data

# Add a random application window to the desktop screenshot
def addApplicationWindow(screen, app_path, fullscreen):
    # Load application screenshot
    app = cv2.imread(app_path, cv2.IMREAD_UNCHANGED)
    
    # Check if we need to add a application head to screenshot, of whether it already contains one
    additional_offset = 0
    if False:
        additional_offset = window_end.shape[0]
    
    # Resize application such that it is not lager than the screenshot
    app = imutils.resize(app, width=min(screen.shape[1] - taskbar_end.shape[1] - border_effect_size * 2, app.shape[1]))
    app = imutils.resize(app, height=min(screen.shape[0] - topbar_start_original.shape[0] - additional_offset - border_effect_size * 2, app.shape[0]))

    # Disable fullscreen mode for applications which would be rescaled too large 
    if app.shape[0] < screen.shape[0] / 1.5 or app.shape[1] < screen.shape[1] / 1.5:
        #print(app_path, " is too small for fullscreen mode!")
        fullscreen = False
 
    if fullscreen:
        # Resize application to fullscreen
        app = cv2.resize(app, (screen.shape[1] - taskbar_end.shape[1], screen.shape[0] - topbar_start_original.shape[0]))
    else:
        # Randomly scale the application, since it's not in fullscreen
        scale = random.uniform(0.7, 1.0)
        app = imutils.resize(app, width=int(app.shape[1] * scale))

    # Generate a random application head if necessary
    if additional_offset > 0:
        head = generateWindowHead(app.shape[1])
        app = np.vstack((head, app))
        
    app = cv2.cvtColor(app, cv2.COLOR_RGB2RGBA)
        
     # Add window effect to application (if not in fullscreen)
    if not fullscreen:
        app = addWindowBorderEffect(app)

    # Choose random position for application
    x1 = random.randint(taskbar_end.shape[1], screen.shape[1] - app.shape[1])
    x2 = x1 + app.shape[1]
    
    y1 = random.randint(topbar_start_original.shape[0], screen.shape[0] - app.shape[0])
    y2 = y1 + app.shape[0]

    # Add application to screenshot
    screen = insertTransparentImage(screen, app, x1, x2, y1, y2)
    
    data = {"name": app_path, "fullscreen": fullscreen, "x1": x1, "x2": x2, "y1": y1, "y2": y2}
    
    return screen, fullscreen, data
```

```python
# Create a random desktop shortcut
def createShortcut(icon_name, icon_size):
    # Vocabulary for shortcut text
    words = ["lorem", "ipsum", "dolor", "sit", "amet", "consetetur", "sadipscing", "elitr", "sed", 
         "diam", "nonumy", "eirmod", "tempor", "invidunt", "ut", "labore", "et", "dolore", "magna", 
         "aliquyam", "erat", "sed", "diam", "voluptua", "at", "vero", "eos", "et", "accusam", "et", 
         "justo", "duo", "dolores", "rebum"]

    # Padding of the shortcut
    icon_within_border = int(icon_size * 0.26)

    # Height of the icon
    icon_image_max_height = int(icon_size * 0.8) - icon_within_border
    
    # Height of the shortcut text
    icon_text_max_height = icon_size - icon_image_max_height - icon_within_border

    # Get the shortcut font
    fontsize = int(icon_within_border / 2.0)
    font = ImageFont.truetype(path_to_font, fontsize)
    
    # Load the shortcut icon
    icon = cv2.imread(os.path.join(path_to_icons, icon_name), cv2.IMREAD_UNCHANGED)

    # Max possible width of icon
    max_width = min(icon_size - 2*icon_within_border, icon.shape[1])
    icon = imutils.resize(icon, width=max_width)

    # Max possible height of icon
    max_height = min(icon_image_max_height, icon.shape[0])
    icon = imutils.resize(icon, height=max_height)

    # Center icon 
    x = int((icon_size - icon.shape[1]) / 2)
    y = icon_within_border + int((icon_image_max_height - icon.shape[0]) / 2)

    # Add the icon
    image = np.zeros((icon_size, icon_size, 4), np.uint8)
    image[y:y+icon.shape[0], x:x+icon.shape[1]] = icon

    # Generate the random text
    text = " ".join(random.choices(words, k=random.randint(1,3)))
    
    # Add text
    image = addShortcutText(image, text, font, icon_within_border, icon_text_max_height, icon_image_max_height)
    
    return image
```

```python
def fillDesktop(screen):
    # ==> Desktop Icons:
    
    coinflip = random.randint(0, 2)
    
    if coinflip == 0:
        # Select random number of desktop shortcuts
        number_of_desktop_icons = random.randint(6, 35)

        # Load desktop shortcut icons
        icon_files = os.listdir(path_to_icons)
        icons = random.choices(icon_files, k=number_of_desktop_icons)

        # Desktop shortcut vocabulary
        words = ["Lorem", "ipsum", "dolor", "sit", "amet", "consetetur", "sadipscing", "elitr", "sed", 
                 "diam", "nonumy", "eirmod", "tempor", "invidunt", "ut", "labore", "et", "dolore", "magna", 
                 "aliquyam", "erat", "sed", "diam", "voluptua", "at", "vero", "eos", "et", "accusam", "et", 
                 "justo", "duo", "dolores", "rebum"]

        # Define shortcut size
        icon_size = int(screen.shape[0] / 12)

        # Add shortcuts to desktop
        for i in range(0, len(icons)):
            # Create shortcut
            image = createShortcut(icons[i], icon_size)

            # Select random location for shortcut
            x1 = random.randint(taskbar_end.shape[1] + 10, screen.shape[1] - image.shape[1] - 10)
            x2 = x1 + image.shape[1]

            y1 = random.randint(10 + topbar_start_original.shape[0], screen.shape[0] - image.shape[0] - 10)
            y2 = y1 + image.shape[0]

            # Draw shortcut
            screen = insertTransparentImage(screen, image, x1, x2, y1, y2)
    elif coinflip == 1:
        pass
    else:
        # Define shortcut size
        icon_size = int(screen.shape[0] / 12)

        # Define additional (horizontal) margin between shortcuts
        icon_outside_border = int(icon_size * 0.05)

        # Compute grid size for shortcuts
        row_size = int((screen.shape[0] - topbar_start_original.shape[0]) / icon_size)
        col_size = int((screen.shape[1] - taskbar_end.shape[1] - icon_outside_border) / (icon_size + icon_outside_border))

        # Enlarge horizontal margin between shortcuts such that there is no potential large gap at end
        rest_outside_border = screen.shape[1] - taskbar_end.shape[1] - (col_size * icon_size + (col_size+1) * icon_outside_border)
        horizontal_outside_border = int(rest_outside_border / (col_size + 1)) + icon_outside_border

        # Enlarge vertical margin between shortcuts such that there is no potential large gap at end
        rest_outside_border = screen.shape[0] - topbar_start_original.shape[1] - (row_size * icon_size + (row_size+1) * icon_outside_border)
        vertical_outside_border = int(rest_outside_border / (row_size + 1)) + icon_outside_border

        # Chance that there is a block of shortcuts at the border of the screen
        # The larger the number, the less likely
        border_chance = random.randint(5, 10)

        # Chance that there is a block of shortcuts within the screen
        normal_chance = random.randint(border_chance*2, border_chance*4)

        # Grid matrix which hold where a shortcut is placed (1 shortcut, 0 no shortcut)
        icon_matrix = np.zeros((row_size, col_size))

        # Traverse the border of the grid (counter clockwise)
        steps = [[1, 0], [0, 1], [-1, 0], [0, -1]]

        x = 0
        y = 0
        i = 0

        while True:
            # Check if we hit a corner, if so, change the traverse direction
            if y + steps[i][0] < 0 or y + steps[i][0] == icon_matrix.shape[0]:
                i += 1
            elif x + steps[i][1] < 0 or x + steps[i][1] == icon_matrix.shape[1]:
                i += 1

            # We did one whole turn around the border, stop
            if i == 4:
                break

            # Get next grid field
            x += steps[i][1]
            y += steps[i][0]

            # If there already is an shortcut, continue
            if icon_matrix[y, x] == 1:
                continue

            # Decide if there is a shortcut
            coinflip = random.randint(0, border_chance)

            if coinflip == 0:
                # We place a shortcut. Now decide how many surrounding shortcuts there are

                # Vertical direction
                if steps[i][0] == 1 or (steps[i][0] == 0 and y == 0):
                    # We are at left border (going down) or at the top border (going left)
                    y1 = y
                    y2 = y + random.randint(0, min(2, icon_matrix.shape[0] - y - 1)) + 1
                else:
                    y1 = y - random.randint(0, min(2, y))
                    y2 = y + 1

                # Horizontal direction
                if steps[i][1] == 1 or (steps[i][1] == 0 and x == 0):
                    # We are at bottom border (going right) or at the left border (going down)
                    x1 = x
                    x2 = x + random.randint(0, min(4, icon_matrix.shape[1] - x - 1)) + 1
                else:
                    x1 = x - random.randint(0, min(4, x))
                    x2 = x + 1

                # Fill with 1s (i.e., add block of shortcuts)
                icon_matrix[y1:y2, x1:x2] = 1

        # Select a random number of desktop icons which will be added to the left side of screen (one.after.the.other)
        number_of_icons = random.randint(6, 25)
        for i in range(0, number_of_icons):
            row = int(i / row_size)
            col = i % row_size

            icon_matrix[col, row] = 1

        # Decide for shortcut blocks within the screenshot (not border)
        for y in range(0, icon_matrix.shape[0]):
            for x in range(0, icon_matrix.shape[1]):
                if icon_matrix[y, x] == 1:
                    continue

                # Randomly decide if we add a shortcut
                coinflip = random.randint(0, normal_chance)

                if coinflip == 0:
                    # Decide for size of shortcut block
                    height = y + random.randint(1, min(2, icon_matrix.shape[0] - y))
                    width = x + random.randint(1, min(3, icon_matrix.shape[1] - x))

                    icon_matrix[y:height, x:width] = 1

        #print(icon_matrix)

        # Draw shortcuts to the screenshot
        icon_files = os.listdir(path_to_icons)

        for i in range(0, row_size):
            for j in range(0, col_size):
                if icon_matrix[i, j] == 1:
                    # Select random icon
                    icon_name = random.choice(icon_files)
                    image = createShortcut(icon_name, icon_size)

                    # Compute position
                    x1 =  taskbar_end.shape[1] + j * icon_size + (j+1) * horizontal_outside_border
                    x2 = x1 + image.shape[1]

                    y1 = topbar_start_original.shape[0] + i * icon_size + (i+1) * vertical_outside_border
                    y2 = y1 + image.shape[0]

                    # Draw on screenshot
                    screen = insertTransparentImage(screen, image, x1, x2, y1, y2)
 

    # ==> Application Windows:
    applications = []
    
    # Captures if at least one of the windows is in fullscreen
    fullscreen_mode = False
    
    # Decide for a random number of application windows
    number_of_applications = random.randint(0, 5)
    
    # Load all application windows form dataset
    for d in os.listdir(path_to_screenshots):
        new_path = os.path.join(path_to_screenshots, d)
        
        if os.path.isdir(new_path):
            for img in os.listdir(new_path):
                applications.append(os.path.join(new_path, img))
    
    # Sample random screenshots
    app_paths = []
    if len(applications) >= number_of_applications:
        # Only do this if we have provided enough application screenshots
        app_paths = random.sample(applications, k=number_of_applications)

    # Load webpage screenshots from dataset
    websites = os.listdir(os.path.join(path_to_websites, "large"))
    
    # Sample random webpages
    website_names = random.sample(websites, k=number_of_applications)

    data = []
    
    # Draw windows
    for i in range(0, number_of_applications):
        coinflip = random.randint(0, 4)
    
        # Decide if fullscreen or not
        if coinflip == 0:
            fullscreen = True
        else:
            fullscreen = False
        
        # Decide if webpage or application 
        coinflip = random.randint(0, 1)
        
        if coinflip == 0 and len(app_paths) > 0:
            screen, f, d = addApplicationWindow(screen, app_paths[i], fullscreen)
        else:
            screen, f, d = addBrowerWindow(screen, website_names[i], fullscreen)
        
        # Update whether one window was in fullscreen mode
        fullscreen_mode = fullscreen_mode or f
        
        data.append(d)

    # ==> Mouse Icon:

    x_offset = random.randint(0, screen.shape[1] - mouse_icon.shape[1])
    y_offset = random.randint(0, screen.shape[0] - mouse_icon.shape[0])

    # Get a random position for mouse icon
    y1 = random.randint(0, screen.shape[0] - mouse_icon.shape[0])
    y2 = y1 + mouse_icon.shape[0]

    x1 = random.randint(taskbar_end.shape[1], screen.shape[1] - mouse_icon.shape[1])
    x2 = x1 + mouse_icon.shape[1]

    # Draw mouse icon
    #screen = insertTransparentImage(screen, mouse_icon, x1, x2, y1, y2)
    
    return screen, fullscreen_mode, data
```
